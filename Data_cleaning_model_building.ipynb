```python
import pandas as pd 
import numpy as np 
import os
import streamlit as st 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
print("All packages imported")
```

```python
data = pd.read_csv(os.path.join("data_assets","employee_data.csv"))
data
```

```python
education_dummies = pd.get_dummies(data.EducationField)
education_dummies
```

```python
travel_dummies = pd.get_dummies(data.BusinessTravel)
travel_dummies
```

```python
department_dummies = pd.get_dummies(data.Department)
department_dummies
```

```python
gender_dummies = pd.get_dummies(data.Gender)
gender_dummies
```

```python
merged = pd.concat([data,education_dummies,department_dummies,travel_dummies,gender_dummies], axis='columns')
merged
```

```python
updated = merged.drop(['Department','EducationField','BusinessTravel','Non-Travel','Sales','Technical Degree','Other','Male'],axis='columns')
updated.head(5)
```

```python
attrition_le = LabelEncoder()
gender_le = LabelEncoder()
```

```python
updated.Attrition = attrition_le.fit_transform(updated.Attrition)
updated.Female = gender_le.fit_transform(updated.Female)
updated
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
X = updated.drop(['Attrition'],axis='columns')
Y = updated.Attrition

X
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```
